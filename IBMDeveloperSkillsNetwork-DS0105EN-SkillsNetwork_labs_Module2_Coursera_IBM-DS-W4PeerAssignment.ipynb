<h3>The code snippet below plots a sine wave between [0,1] radian in 100th increments.</h3>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 1, 0.01)
y = np.sin(2*np.pi*x)
plt.plot(x, y)
plt.xlabel('sample(n)')
plt.ylabel('sin(x)')
plt.show()
```
